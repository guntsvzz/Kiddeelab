```python
import seaborn as sns
import pandas as pd
from sklearn.datasets import load_iris
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load the Iris dataset
iris = load_iris()
# Create a DataFrame
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
# Add the target variable to the DataFrame
df['target'] = iris.target
df['target'] = df['target'].map({i: name for i, name in enumerate(iris.target_names)})
# Print the first few rows of the DataFrame
df.head()
```

```python
df.columns
```

## histplot

```python
sns.histplot(df["sepal length (cm)"],color='darkorange') 
```

```python
sns.histplot(y="sepal length (cm)",data=df,color='darkorange') 
```

```python
sns.histplot(x='sepal length (cm)',data=df,hue=df['target']) 
```

## distplot

```python
sns.distplot(df["sepal length (cm)"],color='darkorange') 
```

```python
sns.distplot(df["sepal length (cm)"],hist=False,color='darkorange')
```

## boxplot

```python
sns.boxplot(df["sepal length (cm)"],color='darkorange') 
```

## countplot 

```python
sns.countplot(df['target'])
```

## barplot 

```python
sns.barplot(x='target',y='sepal length (cm)',data=df) 
```

## pointplot 

```python
sns.pointplot(x='target',y='petal width (cm)',data=df,color='darkorange') 
```

## scatterplot 

```python
sns.scatterplot(df["petal length (cm)"],df['petal width (cm)'],color='darkorange') 
```

## lineplot

```python
sns.lineplot(df['petal length (cm)'],df['petal width (cm)'],color='darkorange')
```

## regplot 

```python
sns.regplot(df['petal length (cm)'],df['petal width (cm)'],color='darkorange') 
```
