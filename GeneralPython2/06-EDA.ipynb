```python
## Univariate Analysis
#1. countplot is to plot the discrete x_1 or y
# sns.countplot(df.Status)
sns.countplot(data = df, x='Status')
```

```python
#2. distplot or displt is to plot the continous x_1 or y
# for column in df.columns.tolist():
#     sns.displot(data = df, x=column)
sns.displot(df.Alcohol)
#exercise: try displot for another continuous variable
```

```python
df.columns = df.columns.str.strip()
df.columns
```

```python
#3. boxplot is to plot discrete x_1 feautres vs continuous x_2/y
sns.boxplot(x = df['Status'], y = df["Life expectancy"])
```

```python
#4. scatterplot is to plot continuous x_1 vs continuous x_2/y
sns.scatterplot(x = df['Alcohol'], y = df["Life expectancy"], hue = df['Status'])
```
